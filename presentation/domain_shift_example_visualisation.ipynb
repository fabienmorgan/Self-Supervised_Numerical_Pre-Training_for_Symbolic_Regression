```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def true_function(x):
    return np.exp(-x) * np.sin(np.pi * x)

# Define the predicted function (as an example of a poor prediction)
def predicted_function(x):
    return 0.5 * np.exp(-x) * np.sin(2 * np.pi * x)

# Generate data points for in-domain (training range)
x_train = np.linspace(0, 2, 20)
y_train = true_function(x_train)

# Generate data points for out-of-domain (inference range)
x_infer = np.linspace(2, 4, 20)
y_infer = true_function(x_infer)

# Generate smooth curve for true function
x_smooth = np.linspace(0, 4, 400)
y_smooth = true_function(x_smooth)
y_predicted_smooth = predicted_function(x_smooth)

# Create the plot
plt.figure(figsize=(14, 10))

# In-domain plot
plt.subplot(2, 1, 1)
plt.plot(x_smooth, y_smooth, 'g--', label='True function')
plt.plot(x_train, y_train, 'bx', label='Training data')
plt.axvspan(0, 2, color='gray', alpha=0.3)
plt.axvline(0, color='gray', linestyle='--')
plt.axvline(2, color='gray', linestyle='--')
plt.text(1.8, 0.65, 'Trained Range', color='black', ha='right', va='top', fontsize=12, bbox=dict(facecolor='white', edgecolor='none', alpha=0.5))
plt.legend()
plt.title('Example Training')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-0.3, 0.7)
plt.xlim(0, 4)

# Out-of-domain plot
plt.subplot(2, 1, 2)
plt.plot(x_smooth, y_smooth, 'g--', label='True function')
plt.plot(x_infer, y_infer, 'rx', label='Inference data')
plt.plot(x_smooth, y_predicted_smooth, 'r-', label='Predicted equation')
plt.axvspan(0, 2, color='gray', alpha=0.3)
plt.axvline(0, color='gray', linestyle='--')
plt.axvline(2, color='gray', linestyle='--')
plt.text(1.8, 0.65, 'Trained Range', color='black', ha='right', va='top', fontsize=12, bbox=dict(facecolor='white', edgecolor='none', alpha=0.5))
plt.legend()
plt.title('Example Inference')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-0.3, 0.7)
plt.xlim(0, 4)

plt.tight_layout()

plt.show()
```
