This notebook is used for testing. It checks if all the evaluaion data is complete

```python
import pandas as pd
```

```python
evaluation_path = '../evaluation/'
evaluation_summary_path = f'{evaluation_path}evaluation_summary.csv'

evaluation_summary = pd.read_csv(evaluation_summary_path)
```

```python
evaluation_summary.head()
```

```python
minimum_supports = range(-30, 11, 1)
support_range = 20
models = ['mmsr_se1_oldloss', 'mmsr_se5_oldloss', 'mmsr_se5_newloss', 'nsr']
number_of_samples = 500

for model in models:
    for minimum_support in minimum_supports:
        assert 10 == len(evaluation_summary[(evaluation_summary['Model'] == model) & 
                           (evaluation_summary['Minimum Support'] == minimum_support) & 
                           (evaluation_summary['Number of Samples'] == number_of_samples) &
                           (evaluation_summary['Maximum Support'] == minimum_support + support_range)]), f"Assertion failed for model {model} and minimum support {minimum_support}"
```

```python
evaluation_summary[(evaluation_summary['Model'] == 'mmsr_se5_oldloss') & 
                           (evaluation_summary['Minimum Support'] == 8) & 
                           (evaluation_summary['Number of Samples'] == 500) &
                           (evaluation_summary['Maximum Support'] == 28)]
```
